```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
data = pd.read_csv('ETTh1.csv')
data.head()
```

```python
data['date'] = pd.to_datetime(data['date'])
data.set_index('date', inplace=True)

split_date = pd.Timestamp('2017-06-30 23:00:00')
train_data = data[:split_date]
test_data = data[split_date + pd.Timedelta(hours=1):]

scaler = StandardScaler()
scaler.fit(train_data)

data_scaled = scaler.transform(data)
data_scaled = pd.DataFrame(data_scaled, columns=data.columns, index=data.index)
data_scaled.head()
```

```python
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = pd.DataFrame(data)
    cols, names = list(), list()
    # (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('%s(t-%d)' % (data.columns[j], i)) for j in range(n_vars)]
    # (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i)[['OT']])
        if i == 0:
            names += [('OT(t)')]
        else:
            names += [('OT(t+%d)' % i)]
    agg = pd.concat(cols, axis=1)
    agg.columns = names
    if dropnan:
        agg.dropna(inplace=True)
    return agg

n_hours = 336  
n_features = 7  
n_forecast = 24 
reframed = series_to_supervised(data_scaled, n_hours, n_forecast)

reframed.head()
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt

reframed_short = reframed.iloc[:500, :]  # 数据量太大好像会过

values = reframed_short.values
n_train_hours = 300
train = values[:n_train_hours, :]
test = values[n_train_hours:, :]

n_obs = n_hours * n_features
train_X, train_y = train[:, :n_obs], train[:, -n_forecast:]
test_X, test_y = test[:, :n_obs], test[:, -n_forecast:]

train_X = train_X.reshape((train_X.shape[0], n_hours, n_features))
test_X = test_X.reshape((test_X.shape[0], n_hours, n_features))

model = Sequential()
model.add(LSTM(50, input_shape=(train_X.shape[1], train_X.shape[2])))
model.add(Dense(n_forecast))
model.compile(loss='mae', optimizer='adam')

history = model.fit(train_X, train_y, epochs=10, batch_size=72, validation_data=(test_X, test_y), verbose=2, shuffle=False)

plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='test')
plt.legend()
plt.show()
```
