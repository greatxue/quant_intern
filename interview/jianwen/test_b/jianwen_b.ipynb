```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cf = pd.read_csv('trades_cf.csv', encoding="GBK", skiprows=1, header=0)
cf['open_time'] = pd.to_datetime(cf['open_time'])
cf['cover_time'] = pd.to_datetime(cf['cover_time'])
cf['cover_date'] = pd.to_datetime(cf['cover_time']).dt.date
cf['date'] = pd.to_datetime(cf['open_time']).dt.normalize()

cf.head()
```

```python
cf2309 = pd.read_csv('cf2309.csv', encoding="GBK", skiprows=1, header=0)
cf2309['date'] = pd.to_datetime(cf2309['trade_date'], format='%Y%m%d')
cf2309
```

```python
cf = pd.merge(cf, cf2309[["date", "close"]], on='date', how='left')
cf.head()
```

**Trades sheet**

```python
trades = pd.DataFrame()
trades['Ticker'] = cf.symbol
trades['Open_trade_datetime'] = cf.open_time
trades['Open_trade_price'] = cf.open_price
trades['Open_trade_num'] = np.where(cf.ls==1, cf.volume, -cf.volume)
trades['size'] = cf.size
trades['open_commision'] = cf.fee / 2
trades['Close_trade_datetime'] = cf.cover_time
trades['Close_trade_price'] = cf.cover_price
trades['Close_trade_num'] = np.where(cf.ls==-1, cf.volume, -cf.volume)
trades['close_commision'] = cf.fee / 2
trades
```

**Position sheet**

```python
cf['previous_close'] = cf['close'].shift(1)
cf['open_time'] = pd.to_datetime(cf['open_time'])
cf['cover_time'] = pd.to_datetime(cf['cover_time'])
cf['date'] = np.where(cf['open_time'].dt.hour >= 21,
                      cf['open_time'].dt.floor('D') + pd.Timedelta(days=1),
                      cf['open_time'].dt.floor('D'))
cf['date'] = pd.to_datetime(cf['date'])
cf['if_cover']=np.where(cf.cover_time>cf.date, 0, 1)
```

```python
cf['adjusted_open_price'] = np.where((cf['if_cover'] == 0) & (cf['open_time'].dt.hour >= 21),
                                     cf['previous_close'], 
                                     cf['open_price'])
tmp = cf.groupby(['date', 'ls']).agg(
    cost_price=('adjusted_open_price', 'mean'),
    number=('volume', 'sum')
).reset_index()
tmp
```

```python
pos = pd.merge(tmp, cf2309[["date", "close"]], on='date', how='left')
pos = pos.rename(columns={'close': 'close_price'})
pos = pos[["date", "ls", "cost_price", "close_price", "number"]]
pos.fillna(pos.mean(numeric_only=True)).round(2)
```

**Balance sheet**

```python
actual_initial_balance = 1000000  
contract_multiplier = cf['tick_size']
size = cf['size']
df_pos = pos.copy()
df_cf = cf.copy()

df_cf['prev_close'] = df_cf['close'].shift(1)
df_cf_grouped = df_cf.groupby('date').last().reset_index()
df_pos = df_pos.merge(df_cf_grouped[['date', 'close', 'prev_close']], on='date', how='left')
df_pos['trade_pnl'] = (df_pos['close_price'] - df_pos['cost_price']) * df_pos['ls'] * df_pos['number'] * contract_multiplier
df_pos['hold_pnl'] = (df_pos['close'] - df_pos['prev_close']) * df_pos['ls'] * df_pos['number'] * contract_multiplier
df_pos['trade_cost'] = df_pos['number'] * contract_multiplier * cf.fee / 2
df_pos['total_pnl'] = df_pos['trade_pnl'] + df_pos['hold_pnl'] - df_pos['trade_cost']
```

```python
agg_ops = {
    'trade_pnl': 'sum',
    'hold_pnl': 'sum',
    'trade_cost': 'sum',
    'total_pnl': 'sum',
    'number': 'sum',
    'close': 'mean'  
}

df_daily = df_pos.groupby('date').agg(agg_ops).reset_index()
df_daily['cum_pnl'] = df_daily['total_pnl'].cumsum()
df_daily['market_value'] = df_daily['number'].abs() * size * df_daily['close']

initial_balance = df_daily.iloc[0]['market_value']
df_daily['balance'] = initial_balance + df_daily['cum_pnl']
df_daily['return'] = df_daily['total_pnl'] / df_daily['balance'].shift(1).fillna(initial_balance)
df_daily['net'] = (1 + df_daily['return'].fillna(0)).cumprod()

bal = df_daily[['date', 'trade_pnl', 'hold_pnl', 'trade_cost', 'total_pnl', 'cum_pnl', 'market_value', 'balance', 'return', 'net']]
bal
```
